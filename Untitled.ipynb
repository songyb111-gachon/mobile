```python
#!/usr/bin/env python3
"""
3D Gaussian beam simulation and 2D plane extraction for comparison
"""

import numpy as np
import matplotlib.pyplot as plt
import meep as mp
import math

def run_3d_gaussian_beam():
    """Run 3D MEEP simulation with exact Gaussian beam"""
    
    # Parameters matching the official tutorial
    s = 14  #    ( : m)
    resolution = 50  #   (/m) -   
    dpml = 2  # PML()  (m) -   
    
    # 3D cell (add z dimension)
    cell_size = mp.Vector3(s, s, s)  # 3    (x, y, z)
    
    boundary_layers = [mp.PML(thickness=dpml)]  # PML   
    
    # Beam parameters
    beam_x0 = mp.Vector3(0, 3.0, 0)  #    (   , m)
    rot_angle = 0  #    ( )
    beam_kdir = mp.Vector3(0, 1, 0).rotate(
        mp.Vector3(0, 0, 1), math.radians(rot_angle)
    )  #     (+y )
    beam_w0 = 0.8  #    (m) -     
    beam_E0 = mp.Vector3(0, 0, 1)  #    (Ez )
    fcen = 1.0  #   (c/m )
    
    # 3D Gaussian beam source (exact solution)
    sources = [
        mp.GaussianBeamSource(
            src=mp.ContinuousSource(fcen),  #   (: fcen)
            center=mp.Vector3(0, -0.5 * s + dpml + 1.0, 0),  #   (x, y, z)
            size=mp.Vector3(s, 0, s),  #   (2D : x-z, y=0)
            beam_x0=beam_x0,  #   
            beam_kdir=beam_kdir,  #   
            beam_w0=beam_w0,  #   
            beam_E0=beam_E0,  #   
        )
    ]
    
    sim = mp.Simulation(
        resolution=resolution,  #  
        cell_size=cell_size,  #   
        boundary_layers=boundary_layers,  #   (PML)
        sources=sources,  #  
        dimensions=3,  #   (3D)
    )
    
    print("Running 3D MEEP simulation...")
    sim.run(until=20)  #    ( )
    
    # Extract 2D plane at z=0 (middle of the cell)
    center = mp.Vector3(0, 0, 0)  #    (x, y, z)
    region = mp.Vector3(s, s, 0)  #    (  , PML )
    
    ez_data = sim.get_array(center=center, size=region, component=mp.Ez)  # Ez  
    
    print(f"3D MEEP data shape: {ez_data.shape}")
    print(f"3D MEEP maximum intensity: {np.max(np.abs(ez_data)**2):.6e}")
    
    # Save the simulation results
    np.save("ez_3d_gaussian_beam.npy", ez_data)
    print("3D MEEP simulation data saved as 'ez_3d_gaussian_beam.npy'")
    
    # Also save simulation parameters
    params = {
        's': s,  #  
        'resolution': resolution,  # 
        'dpml': dpml,  # PML 
        'beam_w0': beam_w0,  #   
        'fcen': fcen,  #  
        'beam_x0': [0, 3.0, 0],  #   
        'source_center': [0, -0.5*s + dpml + 1.0, 0]  #   
    }
    np.save("3d_simulation_params.npy", params)
    print("3D simulation parameters saved as '3d_simulation_params.npy'")
    
    return ez_data, s, dpml, beam_w0, fcen

def theoretical_gaussian_beam_intensity(x, y, w0, wavelength, source_center, beam_focus_relative):
    """Theoretical Gaussian beam intensity calculation
    
    Parameters:
    x, y:    (m)
    w0:    (m) -     
    wavelength:  (m)
    source_center:    [x, y] (m)
    beam_focus_relative:     [x, y] (m)
    """
    
    def rayleigh_range(w0, wavelength):
        """Rayleigh   -     """
        return np.pi * w0**2 / wavelength  # zR = w/
    
    def beam_radius(z, w0, zR):
        """z    """
        return w0 * np.sqrt(1 + (z / zR)**2)  # w(z) = w(1 + (z/zR))
    
    # Absolute position of beam focus (    )
    beam_focus_absolute = source_center + beam_focus_relative  #    
    
    # Distance from beam focus along propagation direction (y-axis)
    #  +y  ,   y - focus_y
    z_prop = y - beam_focus_absolute[1]  #  (y) 
    zR = rayleigh_range(w0, wavelength)  # Rayleigh  
    
    # Beam radius at propagation distance z
    w_z = beam_radius(z_prop, w0, zR)  #   z  
    
    # Transverse distance from beam axis (x direction)
    r_squared = (x - beam_focus_absolute[0])**2  #     
    
    # Gaussian intensity distribution
    intensity = (w0 / w_z)**2 * np.exp(-2 * r_squared / w_z**2)  #   
    
    return intensity

def load_3d_simulation_data():
    """Load previously saved 3D simulation data
    
    Returns:
    ez_data: 3D MEEP  Ez  
    s:   (m)
    dpml: PML  (m)
    beam_w0:    (m)
    fcen:  
    """
    try:
        ez_data = np.load("ez_3d_gaussian_beam.npy")  # Ez   
        params = np.load("3d_simulation_params.npy", allow_pickle=True).item()  #   
        
        print("Loaded 3D MEEP simulation data from file")
        print(f"Data shape: {ez_data.shape}")
        print(f"Maximum intensity: {np.max(np.abs(ez_data)**2):.6e}")
        
        return ez_data, params['s'], params['dpml'], params['beam_w0'], params['fcen']
    except FileNotFoundError:
        print("No saved 3D simulation data found. Running new simulation...")
        return None  #   None 

def compare_3d_with_theory(use_saved_data=True):
    """Compare 3D MEEP result with theoretical Gaussian beam
    
    Parameters:
    use_saved_data:     (True:   , False:  )
    """
    
    # Try to load saved data first
    if use_saved_data:
        loaded_data = load_3d_simulation_data()
        if loaded_data is not None:
            ez_data_3d, s, dpml, beam_w0, fcen = loaded_data
        else:
            ez_data_3d, s, dpml, beam_w0, fcen = run_3d_gaussian_beam()
    else:
        # Run new simulation
        ez_data_3d, s, dpml, beam_w0, fcen = run_3d_gaussian_beam()
    
    # Calculate intensity
    intensity_meep = np.abs(ez_data_3d)**2  # MEEP    (|Ez|)
    
    # Create coordinate arrays
    nx, ny = intensity_meep.shape  #   (x, y  )
    wavelength = 1.0 / fcen  #   ( = c/f)
    
    # Coordinate arrays (matching MEEP's coordinate system)
    x_range = s  #  x  (PML )
    y_range = s  #  y  (PML )
    x = np.linspace(-x_range/2, x_range/2, nx)  # x  
    y = np.linspace(-y_range/2, y_range/2, ny)  # y  
    X, Y = np.meshgrid(x, y)  # 2D  
    
    # MEEP source and focus positions
    source_center = np.array([0, -0.5*s + dpml + 1.0])  #    [x, y] (m)
    beam_focus_relative = np.array([0, 3.0])  #     [x, y] (m)
    beam_focus_absolute = source_center + beam_focus_relative  #    
    
    print(f"Coordinate ranges: x = [{x.min():.2f}, {x.max():.2f}], y = [{y.min():.2f}, {y.max():.2f}] um")
    print(f"Source center: ({source_center[0]:.2f}, {source_center[1]:.2f}) um")
    print(f"Beam focus (relative): ({beam_focus_relative[0]:.2f}, {beam_focus_relative[1]:.2f}) um")
    print(f"Beam focus (absolute): ({beam_focus_absolute[0]:.2f}, {beam_focus_absolute[1]:.2f}) um")
    
    # Theoretical Gaussian beam calculation with 90-degree clockwise rotation then left-right flip
    # 1: 90  : (x, y) -> (y, -x)
    X_rot = Y   #  X 
    Y_rot = -X  #  Y 
    
    # 2:  : (x, y) -> (-x, y)
    X_rotated = -X_rot  #     X  = -Y
    Y_rotated = Y_rot   # Y   = -X
    
    #       ,   
    intensity_theoretical = theoretical_gaussian_beam_intensity(
        X_rotated, Y_rotated, beam_w0, wavelength, source_center, beam_focus_relative
    )  #      (90  + )
    
    # Normalize
    intensity_theoretical_norm = intensity_theoretical / np.max(intensity_theoretical)  #  
    intensity_meep_norm = intensity_meep / np.max(intensity_meep)  # MEEP  
    
    # Find actual beam center in MEEP data
    meep_center_idx = np.unravel_index(np.argmax(intensity_meep), intensity_meep.shape)  #  
    actual_beam_center_y = y[meep_center_idx[0]]  #    y 
    actual_beam_center_x = x[meep_center_idx[1]]  #    x 
    
    print(f"Actual beam center position: x = {actual_beam_center_x:.2f} um, y = {actual_beam_center_y:.2f} um")
    
    # Create visualization
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))  # 2x3   (15x10 )
    
    extent = [x.min(), x.max(), y.min(), y.max()]  #    [x_min, x_max, y_min, y_max]
    
    # 3D MEEP result (2D slice)
    im1 = axes[0,0].imshow(intensity_meep_norm,  #   ( )
                          extent=extent,  #  
                          cmap='inferno',  #  (---)
                          origin='lower',  # y   (  )
                          interpolation='spline36')  #  
    axes[0,0].set_title('3D MEEP (z=0 slice)')  #  
    axes[0,0].set_xlabel('x (um)')  # x 
    axes[0,0].set_ylabel('y (um)')  # y 
    plt.colorbar(im1, ax=axes[0,0])  #  
    
    # Theoretical Gaussian beam
    im2 = axes[0,1].imshow(intensity_theoretical_norm,  #     ( )
                          extent=extent,  #  
                          cmap='inferno',  # 
                          origin='lower',  # y  
                          interpolation='spline36')  # 
    axes[0,1].set_title('Theoretical Gaussian Beam')  # 
    axes[0,1].set_xlabel('x (um)')  # x 
    axes[0,1].set_ylabel('y (um)')  # y 
    plt.colorbar(im2, ax=axes[0,1])  # 
    
    # Difference
    difference = np.abs(intensity_meep_norm - intensity_theoretical_norm)  #   
    im3 = axes[0,2].imshow(difference,  #   ( )
                          extent=extent,  #  
                          cmap='plasma',  #  (--)
                          origin='lower',  # y  
                          interpolation='spline36')  # 
    axes[0,2].set_title('Difference (|3D MEEP - Theory|)')  # 
    axes[0,2].set_xlabel('x (um)')  # x 
    axes[0,2].set_ylabel('y (um)')  # y 
    plt.colorbar(im3, ax=axes[0,2])  # 
    
    # Log scale plots
    im4 = axes[1,0].imshow(np.log10(intensity_meep + 1e-10),  #   MEEP  ( )
                          extent=extent,  #  
                          cmap='viridis',  #  (--)
                          origin='lower',  # y  
                          interpolation='spline36')  # 
    axes[1,0].set_title('3D MEEP (Log Scale)')  # 
    axes[1,0].set_xlabel('x (um)')  # x 
    axes[1,0].set_ylabel('y (um)')  # y 
    plt.colorbar(im4, ax=axes[1,0])  # 
    
    im5 = axes[1,1].imshow(np.log10(intensity_theoretical + 1e-10),  #    ( )
                          extent=extent,  #  
                          cmap='viridis',  # 
                          origin='lower',  # y  
                          interpolation='spline36')  # 
    axes[1,1].set_title('Theory (Log Scale)')  # 
    axes[1,1].set_xlabel('x (um)')  # x 
    axes[1,1].set_ylabel('y (um)')  # y 
    plt.colorbar(im5, ax=axes[1,1])  # 
    
    # Profile comparison at beam center
    beam_center_y_idx = np.argmin(np.abs(y - actual_beam_center_y))  #   y  
    x_profile_meep = intensity_meep_norm[beam_center_y_idx, :]  # MEEP x   (: y, : x)
    x_profile_theory = intensity_theoretical_norm[beam_center_y_idx, :]  #  x  
    
    axes[1,2].plot(x, x_profile_meep, 'r-', label='3D MEEP', linewidth=2)  # MEEP  ( )
    axes[1,2].plot(x, x_profile_theory, 'b--', label='Theory', linewidth=2)  #  ( )
    axes[1,2].axhline(y=1/np.e**2, color='gray', linestyle=':', alpha=0.7, label='1/e')  # 1/e 
    axes[1,2].set_xlabel('x (um)')  # x 
    axes[1,2].set_ylabel('Normalized Intensity')  # y  ( )
    axes[1,2].set_title(f'Transverse Profile (y={actual_beam_center_y:.2f}um)')  # 
    axes[1,2].legend()  #  
    axes[1,2].grid(True, alpha=0.3)  #   ( 30%)
    
    plt.tight_layout()  #    
    plt.savefig('gaussian_3d_comparison.png', dpi=150, bbox_inches='tight')  #   (150 DPI)
    plt.show()  #  
    
    # Quantitative analysis
    print(f"\n=== 3D MEEP vs Theory Comparison ===")
    
    # Beam size comparison (1/e width)
    def find_beam_width(profile, x_coords):
        """1/e    """
        max_val = np.max(profile)  # 
        threshold = max_val / (np.e**2)  # 1/e 
        indices = np.where(profile >= threshold)[0]  #   
        if len(indices) > 0:
            return x_coords[indices[-1]] - x_coords[indices[0]]  #   
        return 0
    
    beam_width_meep = find_beam_width(x_profile_meep, x)  # MEEP  
    beam_width_theory = find_beam_width(x_profile_theory, x)  #   
    
    print(f"1/e beam size comparison:")
    print(f"  3D MEEP: {beam_width_meep:.3f} um")
    print(f"  Theory: {beam_width_theory:.3f} um")
    print(f"  Theoretical (2w): {2*beam_w0:.3f} um")
    print(f"  Relative error: {abs(beam_width_meep - 2*beam_w0)/(2*beam_w0)*100:.1f}%")
    
    # Correlation and RMS error
    mask = (np.abs(X) <= 3*beam_w0) & (np.abs(Y - actual_beam_center_y) <= 3*beam_w0)  #   3w 
    meep_masked = intensity_meep_norm[mask]  #  MEEP 
    theory_masked = intensity_theoretical_norm[mask]  #   
    
    correlation = np.corrcoef(meep_masked, theory_masked)[0,1]  #  
    rms_error = np.sqrt(np.mean((meep_masked - theory_masked)**2))  # RMS  
    
    print(f"\nNear beam center (3w) statistics:")
    print(f"  Correlation coefficient: {correlation:.4f}")
    print(f"  RMS error: {rms_error:.6f}")
    
    # Gaussian beam parameters
    wavelength = 1.0 / fcen  #  (m)
    zR = np.pi * beam_w0**2 / wavelength  # Rayleigh  (m)
    divergence = wavelength / (np.pi * beam_w0)  #  ()
    
    print(f"\nGaussian beam parameters:")
    print(f"  Beam waist radius (w): {beam_w0:.2f} um")
    print(f"  Rayleigh range (zR): {zR:.2f} um")
    print(f"  Divergence angle: {divergence*1000:.2f} mrad")

if __name__ == "__main__":
    # Set use_saved_data=False to force new simulation
    # Set use_saved_data=True to use saved data if available
    compare_3d_with_theory(use_saved_data=True)  #    
```
